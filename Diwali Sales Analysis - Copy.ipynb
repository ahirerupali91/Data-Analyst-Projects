```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Downloads/Diwali Sales Data.csv',encoding='utf-8',encoding_errors='ignore')
df.shape
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
####  Cleaning the Data  ######
```

```python
#dropped blank columns
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
df.info()
```

```python
#checked for null values
pd.isnull(df).sum()
```

```python
#Dopped null values
df.dropna(inplace=True)
```

```python
df.shape
```

```python
#Change Data Types
df['Amount'] = df['Amount'].astype(int)
```

```python
df['Amount'].dtype
```

```python
#Rename column
df.rename(columns = {'Marital_Status':'Married/Unmarried'},inplace=True)
```

```python
df.info()
```

```python
#describe method() returns description of the data in the DataFrame (i.e count, std, mean etc)
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
##### Exploratory Data Analysis #####
```

```python
# Gender#
```

```python
df.columns
```

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
```

```python
# Age #
```

```python
df.columns
```

```python
ax = sns.countplot(data = df, x = 'Age Group',hue = 'Gender')

for bars in ax.containers:ax.bar_label(bars)
```

```python
# Total Amount vs Age Group #
```

```python
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

```python
# State #
```

```python
# Total orders from top 10 states

sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
ax = sns.barplot(data = sales_state, x='State', y='Orders')

for bars in ax.containers:ax.bar_label(bars)
```

```python
# Total Amount/sales from top 10 states 
```

```python
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_state, x='State', y='Amount')
```

```python
# Marital Status #
```

```python
ax = sns.countplot(data=df, x='Married/Unmarried')

sns.set(rc={'figure.figsize':(5,5)})

for bar in ax.containers:ax.bar_label(bars)
```

```python
sales_Marital_States = df.groupby(['Married/Unmarried','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(data=sales_Marital_States, x='Married/Unmarried', y='Amount', hue='Gender')
```

```python
##### Occupation ########
```

```python
sns.set(rc={'figure.figsize':(18,5)})

ax = sns.countplot(data=df, x='Occupation')
for bars in ax.containers:ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data=sales_state, x='Occupation', y='Amount')

```

```python
#### Product Category ######
```

```python
sns.set(rc={'figure.figsize':(20,5)})

ax = sns.countplot(data=df,x='Product_Category')

for bars in ax.containers:ax.bar_label(bars)
```

```python
Product_category_sales = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data=Product_category_sales, x='Product_Category',y='Amount')
```

```python
### Top 10 sold products as per Product ID ####
```

```python
fig1, ax1 = plt.subplots(figsize=(12,7))

df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
